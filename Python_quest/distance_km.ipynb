```python
import pandas as pd

apt_add = pd.read_csv('/content/addr_lat_lon.csv')
busanstation_add = pd.read_csv('/content/busanstation1csv.csv')
seoulstation_add = pd.read_csv('/content/station_coo.csv')
```

```python
apt_add
```

```python
apt_add1=apt_add.rename(columns={
    'latitude': 'y',
    'longitude': 'x'
})
```

```python
busanstation_add
```

```python
!pip install haversine
from haversine import haversine

# 두 지점 간 거리 계산 함수
def calculate_distance(row1, row2):
    point1 = (row1['y'], row1['x'])
    point2 = (row2['y'], row2['x'])
    return haversine(point1, point2)

# 결과를 저장할 데이터프레임 생성
results = []

# 두 데이터프레임의 모든 지점 쌍에 대해 거리 계산
for idx1, row1 in apt_add1.iterrows():
    for idx2, row2 in busanstation_add.iterrows():
        distance = calculate_distance(row1, row2)
        results.append({
            'full_jibun_addr': row1['full_jibun_addr'],
            'station': row2['station'],
            'distance_km': distance
        })

# 결과를 데이터프레임으로 변환
distance_df = pd.DataFrame(results)
```

```python
distance_df
```

```python
distance_df1 = distance_df[distance_df['distance_km'] < 0.5]
```
